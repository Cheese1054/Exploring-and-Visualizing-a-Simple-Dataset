```python
# Task 1: Exploring and Visualizing the Iris Dataset
# Goal: Load, inspect, and visualize data trends and distributions.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset using pandas (via seaborn's built-in loader)
df = sns.load_dataset('iris')
print("Dataset loaded successfully!")
```

```python
# Print shape, columns, and first few rows
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
df.head()
```

```python
# Use .info() and .describe()
print("--- Info ---")
df.info()
print("\n--- Descriptive Statistics ---")
display(df.describe())
```

```python
# 1. Scatter Plot (Relationships)
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title("Sepal Length vs Width")
plt.show()

# 2. Histogram (Distributions)
plt.figure(figsize=(8,5))
sns.histplot(df['petal_length'], kde=True)
plt.title("Distribution of Petal Length")
plt.show()

# 3. Box Plot (Outliers)
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='species', y='petal_width')
plt.title("Petal Width by Species (Outlier Check)")
plt.show()
```
